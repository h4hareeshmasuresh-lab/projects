## amazon product review

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv("amazon product.csv")
df
```

```python
df.shape
df.columns
df.info()
df.describe()
```

```python
df.isnull().sum()
```

```python
import matplotlib.pyplot as plt

plt.hist(df['Rating'])
plt.title("Rating Distribution")
plt.xlabel("Rating")
plt.ylabel("Count")
plt.show()
```

```python
df['Product Quality'].value_counts()
```

```python
pd.crosstab(df['Battery Life'], df['Rating'])
```

```python
df.groupby('Battery Life')['Rating'].mean().plot(kind='bar')
plt.title("Average Rating by Battery Life")
plt.ylabel("Average Rating")
plt.show()
```

```python
df.groupby('Camera Quality')['Rating'].mean()
```

```python
df['Heating'].value_counts().plot(kind='pie', autopct='%1.1f%%')
plt.title("Heating Issue Distribution")
plt.show()
```

```python
df.groupby('Phone Model')['Rating'].mean().sort_values(ascending=False).head(10)
```

```python
df['Review Text'].value_counts().head(10)
```

```python
sns.heatmap(df.corr(numeric_only=True), annot=True, cmap="coolwarm")
plt.title("Correlation Heatmap")
plt.show()
```
