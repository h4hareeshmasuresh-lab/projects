```python
import pandas as pd
data=pd.read_csv("usedcar.csv")
data
```

```python
from sklearn.preprocessing import LabelEncoder
car_name=LabelEncoder()
prize=LabelEncoder()
model_year=LabelEncoder()
fuel_type=LabelEncoder()
ownership=LabelEncoder()
data['car_name']=car_name.fit_transform(data['car_name'])
data['prize']=prize.fit_transform(data['prize'])
data['model_year']=model_year.fit_transform(data['model_year'])
data['fuel_type']=fuel_type.fit_transform(data['fuel_type'])
data['ownership']=ownership.fit_transform(data['ownership'])
data
```

```python
x=data.drop('prize',axis='columns')
y=data.prize
```

```python
model = linear_model.LinearRegression()
model.fit(x,y)
```

```python
model.predict(x)
```
